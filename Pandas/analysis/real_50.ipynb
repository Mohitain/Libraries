```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

n = 1000

data = {
    "OrderID": range(1001, 1001+n),
    "Customer": np.random.choice(
        ["Riya","Aman","Karan","Neha","Rahul","Priya","Ankit","Simran"], n),
    
    "City": np.random.choice(
        ["Delhi","Mumbai","Bangalore","Pune","Chennai","Hyderabad"], n),
    
    "Product": np.random.choice(
        ["Laptop","Mouse","Keyboard","Monitor","Headphones"], n),
    
    "Quantity": np.random.randint(1,5,n),
    
    "Price": np.random.choice(
        [500,800,1500,3000,7000,50000,60000], n),
    
    "PaymentMethod": np.random.choice(
        ["UPI","Card","Cash","NetBanking"], n),
    
    "OrderDate": pd.date_range(
        start="2024-01-01", periods=n, freq="D")
}
```
